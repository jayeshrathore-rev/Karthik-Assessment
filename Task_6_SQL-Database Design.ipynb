Inserting Into Alias Table

```python
import pandas as pd 

# Load the Alias data from the Excel file
alias_data = pd.read_excel('Property & Alias Info.xlsx', sheet_name='ALIAS INFO')

# Rename the column '#' to 'Alias_ID' for consistency with the SQL table
alias_data.rename(columns={'#': 'Alias_ID'}, inplace=True)

# Step 1: Replace NaN/None and empty strings with None (which maps to SQL NULL)
alias_data = alias_data.replace({np.nan: None, 'nan': None, 'NaT': None, '': None})

# SQL query to insert data into the Alias table
insert_query = """
INSERT INTO Alias (Alias_ID, Alias_Name, Phone_Number, Email, Email_Passwords, Area_Code)
VALUES (%s, %s, %s, %s,%s, %s)
"""

# Function to insert row and handle errors
def insert_row(row):
    try:
        cursor.execute(insert_query, (
            row['Alias_ID'],           # Alias ID
            row['ALIAS NAME'],         # Alias Name
            row['PHONE NUMBER'],       # Phone Number
            row['EMAIL'],              # Email
            row['EMAIL PASSWORDS'],    # Email Passwords
            row['AREA CODE']           # Area Code
        ))
    except (IntegrityError, DataError, ValueError) as e:
        print(f"Error: {e} - Row: {row}")

# Loop through the DataFrame and insert data
for _, row in alias_data.iterrows():
    insert_row(row)

# Commit the transaction to save the data
conn.commit()


print("Data inserted successfully into the Alias table.")
```

Inserting Raw email Logs

```python
import os

# Define the folder path containing the email data CSV files
folder_path = 'email_data'

# Create an empty list to hold DataFrames
dataframes = []

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        # Read the CSV file
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        
        # Append the DataFrame to the list
        dataframes.append(df)

# Combine all DataFrames into a single DataFrame
combined_data = pd.concat(dataframes, ignore_index=True)

combined_data['ID'] = range(1 , len(combined_data) +1 )

# Step 1: Replace NaN/None and empty strings with None (which maps to SQL NULL)
combined_data = combined_data.replace({np.nan: None, 'nan': None, 'NaT': None, '': None})

# Define the SQL insert query
insert_query = """
    INSERT INTO raw_email_logs (Email_ID, Subject, Date_Time, Body, Attachments, Alias_ID , ID)
    VALUES (%s, %s, %s, %s, %s, %s , %s)
"""

# Function to insert row and handle errors
def insert_row(row):
    try:
        cursor.execute(insert_query, (
            row['Email ID'],
            row['Subject'],
            row['Date Time'],
            row['Body'],
            row['Attachments'],
            row['Alias ID'] , 
            row['ID']
        ))
    except (IntegrityError, DataError, ValueError) as e:
        print(f"Error: {e} - Row: {row}")

# Loop through the DataFrame and insert data
for _, row in combined_data.iterrows():
    insert_row(row)

# Commit the transaction to save the data
conn.commit()

print("Data inserted successfully into raw_email_logs table.")
```

Inserting Into Classified Follow Up Data

```python
# Load the CSV file
file_path = 'Classified_Follow_Up_Data_V2.csv'
follow_up_data = pd.read_csv(file_path)

# Convert date format (Assuming the date is in 'DD-MM-YYYY' format in the CSV file)
follow_up_data['Date'] = pd.to_datetime(follow_up_data['Date'], format='%d-%m-%Y', errors='coerce').dt.strftime('%Y-%m-%d %H:%M:%S')

follow_up_data['Booking Link'] = follow_up_data['Booking Link'].map({'Yes': True, 'No': False})
follow_up_data['Tour Confirmation'] = follow_up_data['Tour Confirmation'].map({'Yes': True, 'No': False})
follow_up_data['Requests Tour Booking'] = follow_up_data['Requests Tour Booking'].map({'Yes': True, 'No': False})
follow_up_data['Contains Pictures'] = follow_up_data['Contains Pictures'].map({'Yes': True, 'No': False})


# Replace NaN/None and empty strings with None (which maps to SQL NULL)
follow_up_data = follow_up_data.replace({np.nan: None, 'nan': None, 'NaT': None, '': None})

# Define the SQL insert query
insert_query = """
    INSERT INTO classified_follow_up_data (
        Property_ID, Alias_ID, Date, Summary_of_content, Attachments, 
        Type_of_communication, Tour_Confirmation, Booking_Link, 
        Requests_Tour_Booking, Contains_Pictures, Personalized_or_Generalized , ID
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s);
"""
# Function to insert row and handle errors
def insert_row(row):
    try:
        cursor.execute(insert_query, (
            row['Property ID'],
            row['Alias ID'],
            row['Date'],
            row['Summary of content'],
            row['Attachments'],
            row['Type of communication'],
            bool(row['Tour Confirmation']),
            bool(row['Booking Link']),
            bool(row['Requests Tour Booking']),
            bool(row['Contains Pictures']),
            row['Personalized or Generalized'],
            row['ID']
            ))
    except (IntegrityError, DataError, ValueError) as e:
            print(f"Error: {e} - Row: {row}")

# Loop through the DataFrame and insert data
for _, row in follow_up_data.iterrows():
    insert_row(row)

# Commit the transaction to save the data
conn.commit()

print("Data inserted successfully into the classified_follow_up_data table.")
```
